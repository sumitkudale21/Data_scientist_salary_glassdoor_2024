## Salary Estimate

```python
df['Salary Estimate'].unique()
```

```python
# Used Regular Expression For This Task
# Function to clean and format salary values
def clean_salary(salary):
    # Regular expressions to extract salary ranges
    salary_range = re.findall(r'\$?(\d+)[Kk]?\s*-\s*\$?(\d+)[Kk]?', salary)
    if salary_range:
        min_salary, max_salary = salary_range[0]
        return f"${min_salary}K - ${max_salary}K"
    else:
        # Handling 'Employer Provided Salary' or other formats
        employer_provided = re.findall(r'\$?(\d+)[Kk]?', salary)
        if len(employer_provided) == 2:
            return f"${employer_provided[0]}K - ${employer_provided[1]}K"
        elif len(employer_provided) == 1:
            return f"${employer_provided[0]}K"
        else:
            return salary  # return original if no match

# Apply the function to the 'Salary Estimate' column
df['Salary Estimate'] = df['Salary Estimate'].apply(clean_salary)
```

```python
df.head()
```

```python
def clean_salary(salary):
    # Remove 'K' and '$' and convert to integers
    salary = salary.replace('$', '').replace('K', '')
    if '-' in salary:
        low, high = salary.split('-')
        return (int(low) + int(high)) / 2
    else:
        return int(salary)

# Apply the function to the Salary Estimate column
df['Salary Numeric'] = df['Salary Estimate'].apply(clean_salary)

# Convert to numeric and handle missing data
df['Salary Numeric'] = df['Salary Numeric'].apply(lambda x: x * 1000)
df['Salary Numeric'] = pd.to_numeric(df['Salary Numeric'], errors='coerce')
df['Salary Numeric'].fillna(df['Salary Numeric'].median(), inplace=True)

df.head()
```

## Company Name

```python
df['Company Name'].unique()
```

```python
def clean_company_name(company):
    # Remove trailing '\n<rating>' if present
    company = re.sub(r'\n\d+\.\d+$', '', company).strip()
    # Capitalize the first letter of each word
    company = ' '.join(word.capitalize() for word in company.split())
    return company

# Apply the function to the 'Company Name' column
df['Company Name'] = df['Company Name'].apply(clean_company_name)
df.head()
```

## Location

```python
# Function to split location into city and state
def split_location(location):
    if location.lower() == 'remote':
        return location, location  # Keep 'Remote' as it is
    else:
        parts = location.split(',')
        if len(parts) == 2:
            city = parts[0].strip()
            state = parts[1].strip()
            return city, state
        else:
            return location, ''  # Return original location and empty string if split fails

# Apply the function to create new 'City' and 'State' columns
df['City'], df['State'] = zip(*df['Location'].apply(split_location))

# Drop the 'Location' column
df = df.drop(columns=['Location'])
df.head()
```

```python
def clean_Type_Of_Ownership(type):
    if type == 'Company - Private':
        return 'Private'
    elif type == 'Company - Public':
        return 'Public'
    else:
        return "Others"
    
df['Type of ownership'] = df['Type of ownership'].apply(clean_Type_Of_Ownership)
df.head()
```

## Revenue

```python
df['Revenue'].value_counts()
```

```python
def numerical_revenue(revenue):
    if revenue == 'Unknown / Non-Applicable' or revenue == '-1':
        return 0
    elif revenue == 'Less than $1 million (USD)':
        return 0.5
    elif revenue == '$1 to $5 million (USD)':
        return 3
    elif revenue == '$5 to $25 million (USD)':
        return 15
    elif revenue == '$25 to $100 million (USD)':
        return 62.5
    elif revenue == '$100 to $500 million (USD)':
        return 300
    elif revenue == '$500 million to $1 billion (USD)':
        return 750
    elif revenue == '$1 to $5 billion (USD)':
        return 3000
    elif revenue == '$5 to $10 billion (USD)':
        return 7500
    elif revenue == '$10+ billion (USD)':
        return 10000
    else:
        return 0

df['Revenue Numeric'] = df['Revenue'].apply(numerical_revenue)
df.head()
```
